```python
import csv
import datetime

with open('Menu.txt', 'w') as file:
    file.write("* Please Select a Pizza Base:  1: Classic 2: Margarita 3: Turkish Pizza4: PlainPizza * and the sauce of your choice: 11: Olive12: Mushrooms 13: Goat Cheese 14: Meat 15: Onion 16: Corn * Thank you!")
```

```python
class Pizza:
    def __init__(self, description, cost):
        self.description = description
        self.cost = cost
        
    def get_description(self):
        return self.description
    
    def get_cost(self, order_quantity=1, topping_quantity=0):
        return self.cost * order_quantity + self.get_topping_cost() * topping_quantity

    def get_topping_cost(self):
        return 0

class ClassicPizza(Pizza):
    def __init__(self):
        self.description = "Classic Pizza"
        self.cost = 10.99
        
    def get_cost(self):
        return self.cost
    
class MargheritaPizza(Pizza):
    def __init__(self):
        self.description = "Margherita Pizza"
        self.cost = 11.99
        
    def get_cost(self):
        return self.cost
    
class TurkishPizza(Pizza):
    def __init__(self):
        self.description = "Turkish Pizza"
        self.cost = 12.99
        
    def get_cost(self):
        return self.cost
    
class PlainPizza(Pizza):
    def __init__(self):
        self.description = "Plain Pizza"
        self.cost = 9.99
        
    def get_cost(self):
        return self.cost
    
class PizzaDecorator(Pizza):
    def __init__(self, pizza):
        self.component = pizza

    def get_cost(self):
        return self.component.get_cost() + self.cost

    def get_description(self):
        return self.component.get_description() + " " + self.description

        
class Olive(PizzaDecorator):
    def __init__(self, pizza):
        self.description = "Olive"
        self.cost = 0.99
        PizzaDecorator.__init__(self, pizza)
        
    def get_cost(self):
        return self.cost + PizzaDecorator.get_cost(self)
    
class Mushroom(PizzaDecorator):
    def __init__(self, pizza):
        self.description = "Mushroom"
        self.cost = 0.99
        PizzaDecorator.__init__(self, pizza)
        
    def get_cost(self):
        return self.cost + PizzaDecorator.get_cost(self)
class GoatCheese(PizzaDecorator):
    def __init__(self, pizza):
        self.description = "Goat Cheese"
        self.cost = 1.99
        PizzaDecorator.__init__(self, pizza)
        
    def get_cost(self):
        return self.cost + PizzaDecorator.get_cost(self)

class Meat(PizzaDecorator):
    def __init__(self, pizza):
        self.description = "Meat"
        self.cost = 2.50
        PizzaDecorator.__init__(self, pizza)
        
    def get_cost(self):
        return self.cost + PizzaDecorator.get_cost(self)
class Onion(PizzaDecorator):
    def __init__(self, pizza):
        self.description = "Onion"
        self.cost = 1.20
        PizzaDecorator.__init__(self, pizza)
        
    def get_cost(self):
        return self.cost + PizzaDecorator.get_cost(self)
class Corn(PizzaDecorator):
    def __init__(self, pizza):
        self.description = "Corn"
        self.cost = 0.99
        PizzaDecorator.__init__(self, pizza)
        
    def get_cost(self):
        return self.cost + PizzaDecorator.get_cost(self)
    

# Pizza açıklamaları için bir sözlük tanımlayın
pizza_descriptions = {
    "1": "Classic Pizza: Pizza made only with mozzarella cheese and tomato sauce.",
    "2": "Margherita Pizza: Pizza prepared with tomato sauce, mozzarella cheese and fresh basil leaves.",
    "3": "Turkish Style Pizza: Pizza prepared with sausage, sausage, ham, pepper, tomato and mozzarella cheese.",
    "4": "Plain Pizza: Pizza prepared with tomato sauce only."
}
```

```python
# Pizza seçeneklerini ve açıklamalarını yazdır
print("Please select a pizza base:")
print("1: Classic Pizza: Pizza made only with mozzarella cheese and tomato sauce.")
print("2: Margherita Pizza: Pizza prepared with tomato sauce, mozzarella cheese and fresh basil leaves.")
print("3: Turkish Style Pizza: Pizza prepared with sausage, sausage, ham, pepper, tomato and mozzarella cheese.")
print("4: Plain Pizza: Pizza prepared with tomato sauce only.")

# Pizza seçimi yapın ve ilgili açıklamayı yazdırın
pizza_type = input("Please select a pizza base: 1:Classic Pizza 2: Margherita Pizza 3: Turkish Pizza 4: Plain Pizza ")
if pizza_type in pizza_descriptions:
    print(pizza_descriptions[pizza_type])
    pizza = ClassicPizza() if pizza_type == "1" else MargheritaPizza() if pizza_type == "2" else TurkishPizza() if pizza_type == "3" else PlainPizza()
else:
    print("You have made an invalid choice.")

# Seçilen sosların özelliklerini alın
toppings = []
while True:
    topping_choice = input("11: Olive 12: Mushroom 13: GoatCheese 14: Meat: 15: Onion 16: Corn or press 'q' to exit: ")
    if topping_choice == 'q':
        break
    elif topping_choice in ['11', '12', '13', '14', '15', '16']:
        if topping_choice == '11':
            pizza = Olive(pizza)
        elif topping_choice == '12':
            pizza = Mushroom(pizza)
        elif topping_choice == '13':
            pizza = GoatCheese(pizza)
        elif topping_choice == '14':
            pizza = Meat(pizza)
        elif topping_choice == '15':
            pizza = Onion(pizza)
        elif topping_choice == '16':
            pizza = Corn(pizza)
        toppings.append(topping_choice)
    else:
        print("You have made an invalid choice.")

# Toplam tutarı hesaplayın ve yazdırın
total_amount = pizza.get_cost()
print("You have ordered a", pizza.get_description(), "pizza with the following toppings:")
for topping in toppings:
    if topping == '11':
        print("- Olive")
    elif topping == '12':
        print("- Mushroom")
    elif topping == '13':
        print("- Goat Cheese")
    elif topping == '14':
        print("- Meat")
    elif topping == '15':
        print("- Onion")
    elif topping == '16':
        print("- Corn")
print("Total amount:", total_amount)


# Müşteri bilgilerini alın
name = input("Name: ")
id_number = input("ID Number: ")
credit_card_number = input("Your Credit Card Number: ")
credit_card_password = input("Credit Card Password: ")

# Sipariş tarihini alın
now = datetime.datetime.now()
order_date = now.strftime("%Y-%m-%d %H:%M:%S")

# Verileri Orders_Database.csv dosyasına yazın
with open("Orders_Database.csv", mode="a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow([name, id_number, credit_card_number, credit_card_password, pizza.get_description(), total_cost, order_date])

# Toplam tutarı göster
print("Your order is complete. Thanks!")
```
